# Task 3 Implementations

```python
import numpy as np
from scipy.optimize import minimize

# Function that was given 
def func(x):
    return 1/100 * (x[0]**2 + x[1]**2) - 0.5 * (np.cos(3*x[0]-3*x[1]) + np.cos(3*x[0]+3*x[1]))

# Jakobi Matrix since we need it for the newton method
def jakobi(x):
    derx = 0.02 * x[0] + 1.5 * np.sin(3*x[0]-3*x[1]) + 1.5 * np.sin(3*x[0]+3*x[1])
    dery = 0.02 * x[1] + 1.5 * np.sin(3*x[0]+3*x[1]) - 1.5 * np.sin(3*x[0]-3*x[1])
    return [derx, dery]

starting_points = [[1, 1], [-2, 2], [1.5, 1.5], [3,3], [-3,-3]]

# Run through starting points and get the results
for s in starting_points:
    bfgs = minimize(func, s, method='BFGS')
    newton = minimize(func, s, jac=jakobi, method='Newton-CG')
    
    print(f"Starting Point: {s}")
    print("BFGS:")
    print(f"Solution: {bfgs.x}, Minimum Value: {bfgs.fun}")
    print("Newton:")
    print(f"Solution: {newton.x}, Minimum Value: {newton.fun}")

    iterations = bfgs['nit'] 
    evaluations = bfgs['nfev'] 
    iterations2 = newton['nit'] 
    evaluations2 = newton['nfev'] 

    print("Number of Iterations for BFGS:", iterations)
    print("Number of Function Evaluations for BFGS:", evaluations)
    print("Number of Iterations for Newton:", iterations2)
    print("Number of Function Evaluations for Newton:", evaluations2)
```

# 3D Plot of the function

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate data for plotting
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = func([X, Y])

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(X, Y, Z, edgecolor='k')

ax.set_xlabel('x vals')
ax.set_ylabel('y vals')
ax.set_zlabel('function val')
ax.set_title('3D Plot of the Function')
ax.view_init(elev=0, azim=15)

fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10)

plt.show()
```

```python
from scipy.optimize import minimize

def func2(x): 
    return (x[0]-1)**2 + 100*(x[0]**2-x[1])**2

starting_points = [[1, 1], [-2, 2], [1.5, 1.5]]

# Run through starting points and get the results
for s in starting_points:
    nelder = minimize(func2, s, method='Nelder-Mead')
    cg = minimize(func2, s, method='CG')
    
    print(f"Starting Point: {s}")
    print("Nelder:")
    print(f"Solution: {nelder.x}, Minimum Value: {nelder.fun}")
    print("CG:")
    print(f"Solution: {cg.x}, Minimum Value: {cg.fun}")

    iterations = nelder['nit'] 
    evaluations = nelder['nfev'] 
    iterations2 = cg['nit'] 
    evaluations2 = cg['nfev'] 

    print("Number of Iterations for Nelder:", iterations)
    print("Number of Function Evaluations for Nelder:", evaluations)
    print("Number of Iterations for CG:", iterations2)
    print("Number of Function Evaluations for CG:", evaluations2)
```
