```python
import math
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return (x**4/12) - ((x*math.cos(8*x))/64) + math.sin(8*x)/256

def f2(x):
    return (1/3)*x**3-(1/64)*math.cos(8*x)-(1/4)*x*math.sin(8*x)

def f3(x):
    return x**2 - (1/8)*math.sin(8*x)-2*x*math.cos(8*x)

bound = np.arange(-10, 11, 1)
f1val = [f(x) for x in bound]
f2val = [f2(x) for x in bound]
f3val = [f3(x) for x in bound]

plt.plot(bound, f1val, c="red", label="f(x)")
plt.plot(bound, f2val, c="blue", label="f'(x)")
plt.plot(bound, f3val, c="green", label="f''(x)")
plt.legend()
```

```python
from scipy.optimize import linprog
import warnings

warnings.filterwarnings("ignore")

bounds = [(-5, 2), (0, 7), (-18, 1), (-1, 1)]

def f(x):
    x1, x2, x3, x4 = x
    return 2*x1 + 9*x2 - 3*x3 + x4

x0 = [0,0,0,0]

f = [2, 9, -3, 1]

result = linprog(f, bounds=bounds, x0=x0)

print("Optimal values for function:", result.x)
print("Optimal ouput:", result.fun)
```
