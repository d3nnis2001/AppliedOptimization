## 1 Warm Up

```python
import math

def f(x, y):
    return ((1/100)*(x**2 + y**2) - (1/2)*(math.cos(3*x - 3*y) + math.cos(3*x + 3*y)))

def g(x, y):
    return (x - 1)**2 + 100*(x**2 - y)**2
```

### a)

```python
import matplotlib.pyplot as plt
import numpy as np

import matplotlib.cbook as cbook
import matplotlib.cm as cm
from matplotlib.patches import PathPatch
from matplotlib.path import Path

delta = 0.025
x = y = np.arange(-10, 10, delta)
X, Y = np.meshgrid(x, y)
Z = np.vectorize(f)(X, Y)

fig, ax = plt.subplots()
im = ax.imshow(Z, interpolation='bilinear', cmap=cm.RdYlGn,
               origin='lower', extent=[-10, 10, -10, 10])
```

```python
delta = 0.025
x = y = np.arange(-3, 3, delta)
X, Y = np.meshgrid(x, y)
Z = np.vectorize(g)(X, Y)

fig, ax = plt.subplots()
im = ax.imshow(Z, interpolation='bilinear', cmap="viridis",
               origin='lower', extent=[-3, 3, -3, 3])
```

```python
import random
from scipy.optimize import minimize

# Random Point between -10 and 10
p0 = [random.uniform(-10, 10), random.uniform(-10, 10)]

res = minimize(f, p0[0], p0[1], method="Nelder-Mead")
print(res)
```
